# Building a ML Model using the San Fransisco Crime Dataset

```python
from shapely.geometry import Point
from geopandas import GeoDataFrame
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier
```

## Import Weather Data

```python
weather = pd.read_csv('data/weather.csv')
weather['Date'] = pd.to_datetime(weather['Date'])
weather = weather.drop(columns=['id','Max.Dew.PointF','MeanDew.PointF','Min.DewpointF','Max.Sea.Level.PressureIn','Mean.Sea.Level.PressureIn','Min.Sea.Level.PressureIn','Max.VisibilityMiles','Mean.VisibilityMiles','Min.VisibilityMiles','Max.Wind.SpeedMPH','Mean.Wind.SpeedMPH','Max.Gust.SpeedMPH','PrecipitationIn','WindDirDegrees.br...','city','Events','Max.Humidity','Min.Humidity','Max.TemperatureF','Min.TemperatureF'])
weather['season'], unique_seasons = pd.factorize(weather['season'])
print(weather)
print(weather.dtypes)
```

## Importing train.csv

```python
data = pd.read_csv('data/train.csv')
data['Dates'] = pd.to_datetime(data['Dates'])
data['Date_Only'] = data['Dates'].dt.date
data['Date_Only'] = pd.to_datetime(data['Date_Only'])
data = data[data['Y'] <= 50]
print(data.head())
print(data.dtypes)
```

```python
data_merged = pd.merge(data, weather, left_on='Date_Only', right_on='Date', how='left')

print(data_merged)
print(data_merged.dtypes)

data = data_merged
```

## Models Comparison

```python
data['Month'] = data['Dates'].dt.month

data.drop('Dates', axis=1, inplace=True)
data.drop('Date_Only', axis=1, inplace=True)
data.drop('Date', axis=1, inplace=True)

print(data)
```

```python
x_train = data.drop('Category', axis=1)
print(x_train)
y_train = data['Category']
print(y_train)
```

## Random Forest

```python
data_test = data.drop('Category', axis=1)
X_train, X_test, y_train, y_test = train_test_split(data_test, data['Category'], test_size=0.2, random_state=42)

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

predictions = rf.predict(X_test)

acc = accuracy_score(y_test, predictions)
print('Accuracy Random Forest:', acc)
```

```python
rf_precision = precision_score(y_test, predictions, average='weighted')
rf_recall = recall_score(y_test, predictions, average='weighted')
rf_f1 = f1_score(y_test, predictions, average='weighted')
print('Precision Decision Tree:', rf_precision)
print('Recall Decision Tree:', rf_recall)
print('F1-Score Decision Tree:', rf_f1)
```

## Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

dt_predictions = dt.predict(X_test)

dt_accuracy = accuracy_score(y_test, dt_predictions)
print('Accuracy Decision Tree:', dt_accuracy)
```

```python
dt_precision = precision_score(y_test, dt_predictions, average='weighted')
dt_recall = recall_score(y_test, dt_predictions, average='weighted')
dt_f1 = f1_score(y_test, dt_predictions, average='weighted')
print('Precision Decision Tree:', dt_precision)
print('Recall Decision Tree:', dt_recall)
print('F1-Score Decision Tree:', dt_f1)
```
